```python
from brackettools import *
import matplotlib.pyplot as plt
```

```python
d = loadkp('./kpdata/kp17.txt')  
b = [loadbrack('b17.txt')]

kpb = bsim(b,d)
for i in range(0,100000):
    if i%10000 == 0:
        print i
    kpb.sim(method='kpcdf',std=11)

print ''
wdat, wind = kpb.simsum()
```

```python
objects = sorted(wind.keys())
y_pos = np.arange(len(objects))
performance = [wind[key][4] for key in objects]
 
plt.bar(y_pos, performance, alpha=0.5)
plt.xticks(y_pos, objects,rotation=70)
plt.ylabel('Chance of NCAA Tournament [%]')
 
plt.show()
```
